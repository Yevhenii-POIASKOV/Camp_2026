# Clustering aglomerative and DBSCAN

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
```

## Data

```python
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 5, 
                           cluster_std = 0.6, random_state = 0)
colors = np.array([plt.cm.Spectral(val)
          for val in np.linspace(0, 1, len(set(y)))])
plt.figure(figsize=(8,6))

plt.scatter(X[:,0], X[:,1], c= colors[y], s = 20)    
```

## Aglomerative single

```python
clf_single = AgglomerativeClustering(n_clusters = 5, linkage='single')
clf_single.fit(X)
predicted_single = clf_single.fit_predict(X)
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c= colors[predicted_single], marker= '.') 
plt.title ('SINGLE LINKAGE')
```

## Aglomerative complete

```python
clf_complete = AgglomerativeClustering(n_clusters = 5, linkage='complete')
clf_complete.fit(X)
predicted_complete = clf_complete.fit_predict(X)
plt.figure(figsize=(6,6))
plt.scatter(X[:,0], X[:,1], c= colors[predicted_complete], marker= '.') 
plt.title ('COMPLETE LINKAGE')
```

## DBSCAN

```python
clf_dbscan = DBSCAN(eps = 0.3, min_samples = 5)
clf_dbscan.fit(X)
predicted_dbscan = clf_dbscan.labels_
plt.figure(figsize=(6,6))
colors=  np.array(['green','grey', 'orange', 'brown', 'blue', 'yellow'])
colors= np.r_[colors, np.array(['black']*100)]
plt.scatter(X[:,0], X[:,1], c= colors[predicted_dbscan], s= 40, edgecolor = 'black', label='negative', alpha = 0.7)    
```

## K-means albow

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--', color='b')

plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.xticks(k_range)
plt.grid(True)

plt.show()
```

## Висновок

Як можна побачити на графіку, після значення `k=5` значення інерції майже не змінюється.
